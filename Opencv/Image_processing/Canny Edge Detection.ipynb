# Theory
Canny Ege Detection은 이미지에서 edge를 찾아내는 방법 중 하나이다. Canny Edge Detection은 다음과 같은 단계로 이루어진다.
    1. multi-stage algorithm이다.
    2. Noise reduction
        edge detection은 noise에 매우 민감하다. 따라서 noise를 제거하기 위해 Gaussian filter를 사용한다.
    3. Finding Intensity Gradient of the Image
        smoothed image에서 gradient를 찾아내기 위해 Sobel filter를 사용한다.

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('Images/messi5.jpeg',cv.IMREAD_GRAYSCALE)
assert img is not None, "Image not found"
edges = cv.Canny(img,100,200) # img, minVal, maxVal

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
