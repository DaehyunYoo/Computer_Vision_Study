# Goal
- Find Image gradients, edges
- cv.Sobel(), cv.Scharr(), cv.Laplacian()

# Theory
OpenCV에서는 3개의 gradient filter를 제공한다.
- ㄴSobel, Scharr, Laplacian

## 1. Sobel and Scharr Derivatives
Sobel operations는 Gaussian smoothing pluss differentiation operation이다. 이는 edge detection에 매우 유용하다. 사용자는 derivates의 방향을 지정할 수 있다. vertical, horizontal, or both. 또한, kernel size를 지정할 수 있다. (1, 3, 5, 7) 중 하나여야 한다. Scharr는 Sobel보다 더 좋은 결과를 제공한다.

## 2. Laplacian Derivatives
Laplacian은 Sobel의 미분을 이용하여 구할 수 있다. 이는 이미지의 2차 미분을 계산한다. 결과는 edge가 양수인지 음수인지 알 수 없다. 따라서, 결과를 양수로 만들기 위해 절대값을 취한다. 이는 이미지의 edge를 찾는데 매우 유용하다.

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('Images/dave.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None, 'Cannot load image'

laplacian = cv.Laplacian(img, cv.CV_64F) # cv.CV_64F는 결과값을 64bit float로 반환
sobelx = cv.Sobel(img, cv.CV_64F, 1, 0, ksize=5) # x방향으로 미분
sobely = cv.Sobel(img, cv.CV_64F, 0, 1, ksize=5) # y방향으로 미분

plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show()
```
