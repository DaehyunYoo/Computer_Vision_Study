# Goal
morphological operaions란 이미지 처리에서 주로 사용되는 기법 중 하나로, 이미지의 형태를 변형시키는 기법이다. Erosion, Dilation, Opening, Closing이 있다. 

# Theory
Morphological transformations는 이미지의 형태를 변형시키는데 사용되는 기법이다. 이 기법은 이미지의 형태를 변형시키는데 사용되는데, 주로 binary 이미지에서 사용된다. 두 가지 input이 필요한데, original image와 structuring element(kernel)이다. Structuring element는 원하는 변형을 정의하는데 사용된다.

## 1. Erosion
Erosion은 이미지의 형태를 축소시키는데 사용된다. 이미지의 각 pixel에 structuring element를 적용하여, 이미지의 pixel이 structuring element와 일치할 때만 결과 이미지의 pixel을 1로 설정한다. 이때, structuring element는 이미지의 pixel을 완전히 포함하는 형태여야 한다.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('Images/j.png', cv.IMREAD_GRAYSCALE)
assert img is not None, "Could not read the image."

kernel = np.ones((5, 5), np.uint8)
erosion = cv.erode(img, kernel, iterations=1)

plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(erosion, cmap='gray'), plt.title('Erosion')
plt.xticks([]), plt.yticks([])
plt.show()
```

## 2. Dilation
Dilation은 Erosion의 반대이다. pixel element가 structuring element와 일치할 때, 결과 이미지의 pixel을 1로 설정한다. 이렇게 하면 이미지의 흰색 부분은 확장되고, 검은색 부분은 축소된다.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('Images/j.png', cv.IMREAD_GRAYSCALE)
assert img is not None, "Could not read the image."

kernel = np.ones((5, 5), np.uint8)
erosion = cv.dilate(img, kernel, iterations=1)

plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(erosion, cmap='gray'), plt.title('Dilated')
plt.xticks([]), plt.yticks([])
plt.show()
```

## 3. Opening
Opening은 Erosion 다음에 Dilation을 적용하는 것이다. 이 기법은 작은 노이즈를 제거하는데 사용된다.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('Images/j_dot.png', cv.IMREAD_GRAYSCALE)
assert img is not None, "Could not read the image."

kernel = np.ones((5, 5), np.uint8)
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)

plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(opening, cmap='gray'), plt.title('Opening')
plt.xticks([]), plt.yticks([])
plt.show()
```

## 4. Closing
Closing은 Opening의 반대로, Dilation 다음에 Erosion을 적용하는 것이다. 이 기법은 작은 구멍을 채우는데 사용된다.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('Images/j_dot2.png', cv.IMREAD_GRAYSCALE)
assert img is not None, "Could not read the image."

kernel = np.ones((5, 5), np.uint8)
closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)

plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(closing, cmap='gray'), plt.title('Closing')
plt.xticks([]), plt.yticks([])
plt.show()
```

## 5. Morphological Gradient
Morphological Gradient는 Dilation과 Erosion의 차이를 계산하는 것이다. 이 기법은 이미지의 경계를 찾는데 사용된다.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('Images/j.png', cv.IMREAD_GRAYSCALE)
assert img is not None, "Could not read the image."

kernel = np.ones((5, 5), np.uint8)
gradient = cv.morphologyEx(img, cv.MORPH_GRADIENT, kernel)

plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(gradient, cmap='gray'), plt.title('Gradient')
plt.xticks([]), plt.yticks([])
plt.show()
```

## 6. Top Hat
Top Hat은 Opening과 원본 이미지의 차이를 계산하는 것이다. 이 기법은 작은 노이즈를 찾는데 사용된다.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('Images/j.png', cv.IMREAD_GRAYSCALE)
assert img is not None, "Could not read the image."

kernel = np.ones((5, 5), np.uint8)
tophat = cv.morphologyEx(img, cv.MORPH_TOPHAT, kernel)

plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(tophat, cmap='gray'), plt.title('Tophat')
plt.xticks([]), plt.yticks([])
plt.show()
```

## 7. Black Hat
Black Hat은 원본 이미지와 Closing의 차이를 계산하는 것이다. 이 기법은 이미지의 큰 구멍을 찾는데 사용된다.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('Images/j.png', cv.IMREAD_GRAYSCALE)
assert img is not None, "Could not read the image."

kernel = np.ones((5, 5), np.uint8)
blackhat = cv.morphologyEx(img, cv.MORPH_BLACKHAT, kernel)

plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(blackhat, cmap='gray'), plt.title('Blackhat')
plt.xticks([]), plt.yticks([])
plt.show()
```
